```python
def mvp():
##Most Valuable Players
    players_23=fifa23_df.sort_values(by='Value(£)',ascending=False)
    players=players_23.head(20)
    players=players.sort_values(by='Value(£)',ascending=True)
    players_df=players[['Name','Value(£)']].reset_index(drop=True)
    players_df['Value(£)']=players_df['Value(£)']/1e6

    myColor = ['red', 'blue', 'green', 'purple', 'orange' ,'red', 'blue', 'green', 'purple', 'orange','red', 'blue', 'green', 'purple', 'orange','red', 'blue', 'green', 'purple', 'orange']
    plt.figure(figsize=(16, 9))
    plt.barh(players_df['Name'],players_df['Value(£)'],color=myColor )
    plt.subplots_adjust(top=0.938,bottom=0.097,left=0.179,right=0.972,hspace=0.2,wspace=0.2)
    plt.xlabel("Market Value in Million Pounds")
    plt.ylabel("Player")
    plt.title("Most Valuable Players 2023-24")
    plt.show(block=False)
```

```python
def most_valuable_squads():
##Most Valuable Clubs

    clubs_23_df=fifa23_df.groupby(by="Club")['Value(£)'].sum().reset_index()
    clubs_23_df['Value(£)']=clubs_23_df['Value(£)']/1e9
    most_valuable_clubs=clubs_23_df.sort_values(by='Value(£)',ascending=False).head(10)
    most_valuable_clubs=most_valuable_clubs.sort_values(by='Value(£)',ascending=True)
    print(most_valuable_clubs)

    myColor = ['red', 'blue', 'green', 'purple', 'orange']
    plt.figure(figsize=(16, 9))
    plt.barh(most_valuable_clubs['Club'],most_valuable_clubs['Value(£)'] ,color=myColor)
    plt.xlabel("Squad Value in Billion Pounds")
    plt.subplots_adjust(top=0.938,bottom=0.097,left=0.183,right=0.985,hspace=0.2,wspace=0.2)
    plt.ylabel("Club")
    plt.title("Most Valuable Squads")
    plt.show(block=False)
```

```python
def top_20():
    
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b',
            '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#aec7e8', '#ffbb78',
            '#98df8a', '#ff9896', '#6a3d9a', '#ffbb78', '#bcbd22', '#2ca02c',
            '#ff7f0e', '#9edae5']
    top_df=fifa21_df.sort_values(by='Overall',ascending=False).head(20)
    top_df=top_df.sort_values(by='Overall',ascending=True)
    plt.barh(top_df.Name,top_df.Overall,color=colors)
    plt.subplots_adjust(top=0.984,bottom=0.041,left=0.077,right=0.992,hspace=0.2,wspace=0.2)
    plt.plot()
```

```python
def goat():
    # Filter the DataFrame
    cristiano_ronaldo_df = fifa19_df[fifa19_df['Name'] == 'Cristiano Ronaldo']
    lionel_messi_df=fifa19_df[fifa19_df['Name'] == 'L. Messi']

# Assuming you have a list of attribute names
    forward_attribute = ['Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling', 'Curve',
                   'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions',
                   'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
                   'Positioning', 'Vision', 'Penalties', 'Composure']

# Get attribute values for Cristiano Ronaldo
    cristiano_ronaldo_attributes = cristiano_ronaldo_df[forward_attribute].values[0]
    lionel_messi_attributes = lionel_messi_df[forward_attribute].values[0]

# Calculate angles for each attribute
    angles = np.linspace(0, 2 * np.pi, len(forward_attribute), endpoint=False).tolist()
    angles += angles[:1]  # Repeat the first angle to close the plot

# Convert attribute values to a closed loop
    cristiano_ronaldo_attributes = np.append(cristiano_ronaldo_attributes, cristiano_ronaldo_attributes[:1])
    lionel_messi_attributes = np.append(lionel_messi_attributes, lionel_messi_attributes[:1])

# Create the radar plot
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    fig.subplots_adjust(top=.9, bottom=.1, left=0.05, right=0.95)  # Adjust the padding


    ax.plot(angles, cristiano_ronaldo_attributes, color='blue',marker='*')
    ax.plot(angles, lionel_messi_attributes, color='red', marker='o')

    ax.fill(angles, cristiano_ronaldo_attributes, color='blue', alpha=0.25)
    ax.fill(angles, lionel_messi_attributes, color='red', alpha=0.25)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(forward_attribute)
    ax.set_title('The GOAT Debate')
    ax.legend(['Cristiano Ronaldo', 'Lionel Messi'])
    plt.show()
```

```python
def top_20_fwd():
    strings_to_search = ['ST','CF','LW','RW','LF','RF','LS','RS']

    # Extract rows containing the specified strings
    fwd_df = fifa21_df[fifa21_df['Position'].str.contains('|'.join(strings_to_search))].sort_values(by="Overall",ascending=False).head(20).reset_index(drop=True).set_index('ID')
    plt.figure(figsize=(16,9))
    plt.subplots_adjust(top=0.959,bottom=0.043,left=0.1,right=0.991,hspace=0.2,wspace=0.2)
    plt.title("Top 20 Attackers")
    plt.barh(fwd_df['Name'],fwd_df['Overall'],color=colors)
    plt.show()
```

```python
def fwd():
    forward_attribute = ['Name','Finishing' ,'Acceleration']

    forward_attribute_2 = [ 'Finishing' ,'Acceleration']

    strings_to_search = ['ST','CF','LW','RW','LF','RF','LS','RS']

    fwd_df = fifa21_df[fifa21_df['Position'].str.contains('|'.join(strings_to_search))].sort_values(by="Overall",ascending=False).head(20).reset_index(drop=True)
    fwd_df=fwd_df[forward_attribute]
    fwd_df
    
    plt.figure(figsize=(16,9))
    plt.subplots_adjust(top=0.961,bottom=0.061,left=0.032,right=0.972,hspace=0.2,wspace=0.2)
    sns.scatterplot(y=fwd_df.Finishing ,x=fwd_df.Acceleration)
    plt.axhline(y=85, color='red', linestyle='--')
    plt.axvline(x=85, color='green', linestyle='--')
    plt.title('Top attackers')
    
    for i in range(len(fwd_df)):
        plt.annotate(fwd_df['Name'][i], xy=(fwd_df['Acceleration'][i], fwd_df['Finishing'][i]),textcoords="offset points",xytext=(5, 5))
    plt.show(block=False)
```

```python
def top_20_mid():
    strings_to_search = ['DM', 'RDM', 'LDM', 'CM', 'LCM', 'RCM', 'CAM', 'LAM', 'RAM']

    # Extract rows containing the specified strings
    new_df = fifa21_df[fifa21_df['Position'].str.contains('|'.join(strings_to_search))].sort_values(by="Overall",ascending=False).head(20).reset_index(drop=True).set_index('ID')
    plt.figure(figsize=(16,9))
    plt.barh(new_df['Name'],new_df['Overall'],color=colors)
    plt.subplots_adjust(top=0.961,bottom=0.041,left=0.079,right=0.992,hspace=0.2,wspace=0.2)
    plt.title('Top 20 Midfielders')
    plt.show()
```

```python
def cm():
    mid_df=fifa21_df[['Name','Vision','LongPassing']]
    avg=(mid_df['Vision']+ mid_df['LongPassing'])/2
    mid_df['Avg']=avg
    sorted_mid_df=mid_df.sort_values(by='Avg',ascending=False)
    sorted_mid_df=sorted_mid_df.head(100)
    sorted_mid_df=sorted_mid_df.drop_duplicates(subset=['Vision','LongPassing'])
    sorted_mid_df=sorted_mid_df.reset_index(drop=True)
    sorted_mid_df

    print(sorted_mid_df['LongPassing'].mean())
    print(sorted_mid_df['Vision'].mean())

    plt.figure(figsize=(16,9))
    plt.subplots_adjust(top=0.961,bottom=0.061,left=0.032,right=0.992,hspace=0.2,wspace=0.2)
    sns.scatterplot(y=sorted_mid_df.LongPassing ,x=sorted_mid_df.Vision)
    plt.axhline(y=87.2, color='red', linestyle='--')
    plt.axvline(x=87.9, color='green', linestyle='--')
    plt.title('Top Midfielders')
    
    for i in range(len(sorted_mid_df)):
        plt.annotate(sorted_mid_df['Name'][i], xy=(sorted_mid_df['Vision'][i], sorted_mid_df['LongPassing'][i]),textcoords="offset points",xytext=(5, 5))
    plt.show(block=False)
```

```python
def gk():
    gk_attributes=['Name','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes']
    gk_df=fifa21_df[gk_attributes]
    gk_df['GKAverage']=(gk_df['GKDiving']+gk_df[ 'GKHandling']+gk_df['GKKicking']+gk_df['GKPositioning']+gk_df['GKReflexes'])/5
    gk_df
    sorted_gk_df=gk_df.sort_values(by='GKAverage',ascending=False).drop_duplicates(subset=['GKDiving','GKReflexes'])
    sorted_gk_df=sorted_gk_df.head(30).reset_index()
    sorted_gk_df

    plt.figure(figsize=(16,9))
    plt.subplots_adjust(top=0.961,bottom=0.061,left=0.032,right=0.983,hspace=0.2,wspace=0.2)
    sns.scatterplot(y=sorted_gk_df.GKDiving ,x=sorted_gk_df.GKReflexes)
    plt.axhline(y=85.83, color='red', linestyle='--')
    plt.axvline(x=83.86, color='green', linestyle='--')
    plt.title('Top Goalkeepers')

    for i in range(len(sorted_gk_df)):
        plt.annotate(sorted_gk_df['Name'][i], xy=(sorted_gk_df['GKReflexes'][i], sorted_gk_df['GKDiving'][i]),textcoords="offset points",xytext=(5, 5))
    plt.show(block=False)
```
